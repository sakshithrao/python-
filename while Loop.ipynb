```python
# remove the target elements from given list using for loop.
arr=list(map(int, input().split()))
target=int(input())
res=[]
for num in arr:
    if num !=target:
        res.append(num)
print(res)
```

```python
#remove the target elements from given list using while loop
arr = list(map(int, input().split()))
target = int(input())

i = 0
n = len(arr)

while i < n:
    if arr[i] == target:
        print(i)
        break
    i += 1
else:
    print(-1)
```

```python
# removing the target element without using extra space 
# Take list input from user
lst = list(map(int, input("Enter the list elements: ").split()))

# Take target element
target = int(input("Enter the target element to remove: "))

index = 0

for i in range(len(lst)):
    if lst[i] != target:
        lst[index] = lst[i]
        index += 1

# Remove unwanted elements
del lst[index:]

print("List after removing target element:", lst)
```

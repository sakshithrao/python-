### set methods
- Adding Element T0 set
- add(ele):It adds element to set if element to set, if element not exists in set , otherwise nothing .
-update(iterable):It adds all iteable elements to set
  -REMOVE ELEMENTS FROM SET:
- remove(ele):It removes the element from set if element exists in set otherwise nothing .
  -'KEYERROR'
- Discard(ele): It removes the element from set if element exists in set otherwise nothing
-Pop():It remove and returns the element from set
  -SET OPERATIONS
     - Union(anthor_sets):It returns the all common elements from the all sets
-Intersection(another_sets): It returns the common elements from the all sets
-difference(another_sets): It returns these set elements those are not lies in anthor_sets
-symmetric_difference(another_sets): it returns the Non-common elements from all sets
  

```python
s={1,2,3,-5}
s.add(4)
print(s)
s.add(1)# 1 doesn't add to set because set contains 1 
print(s)
s.remove(1)
print(s)

s.discard(1)
print(s)
s.remove(1)#key error if element doesnot exist
print(s)
```

```python
s1={1,2,5,7}
s2={3,4,5,6,7}
s3={7,9}
print(s1.union(s2))
print(s1.union(s2,s3))
print(s1.intersection(s2))
print(s1.intersection(s2,s3))
print(s1.difference(s2))
print(s1.difference(s2,s3))
print(s1.symmetric_difference(s2))
print(s1.symmetric_difference(s2,s3))#symmetric_sifference takes only 1 argument , but we gave 2 so error 



```

```python
s1={1,2,5,7}
s2={3,4,5,6,7}
s3={7,9}
print(s1 |(s2)) # | - union
print(s1|(s2|s3))
print(s1&(s2))#  &-intersection
print(s1&(s2&s3))
print(s1-(s2))# - =difference
print(s1-(s2-s3))
print(s1^(s2))
print(s1^(s2^s3))# ^ = symmetric_ddifference
```

```python
# this methods updates the set object values
s1={1,2,3}
s2={2,3,4,5}

s1.intersection_update(s2)
print(s1)
s1.difference_update(s2)
print(s1,s2)
s1.symmetric_difference_update(s2)
print(s1)
```

```python
# set cheking methods
# is subset(another_set)
#is superset(another_set)
#is disjoint(another_set)

print({1,2,3}.issubset({1,2,3,4,4}))
print({1,2,3,4}.issuperset({1,2}))
print(set().issubset({1}))
print({1,2}.isdisjoint({4,5}))
```

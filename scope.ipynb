#enclosing scope
#this scope can happened through nested functions here a variable defined in outer function , we can access inside the inner functions 
#those variables scope called as  enclosing scope

```python
#Enclosing Scope

def outer(x):
    y=10#enclosed scope
    def inner():
        a=10 # local scope
        z=y+a
        print(f"this is local scope :{z}")
    inner()
    y=y+x
    print(f"this is enclosing scope:{y}")
n=5
outer(n)
```

```python
# Global Scope

x= 10#global scope

def outer():#enclosed
    y=20
    def inner():#local
        z=30
        res=z+y+x
        print(f"this is local scope: {res}")
    res = y + x#30
    inner()
    print(f" this is enclosing scope:{res}")
outer()
```

```python
# update the global scope variable inside the local scope by using 'global' key
x= 10   #global scope   

def outer():
    global x
    x = x + 1
    print("this is local scope val:",x)
outer()
print("this is global scope val:",x)  
```

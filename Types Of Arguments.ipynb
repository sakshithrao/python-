```python
 #types of arguments
# 1. positional arguments  :   the arguments are assigns to parameters as per the positional arguments

def sub(x:int, y:int)->int:
    c=x-y
    return c
#function calling block

a , b = 10,  20

res =sub(a,b)
print(res)
print(sub(b,a))
```

```python
#types of arguments
# 2. Key word  arguments  : 

def sub(x:int, y:int)->int:
    c=x-y
    return c
#function calling block

a , b = 10,  20

print(sub(x = a , y=b))
print(sub(y = b , x =a))
```

```python
# 2 numbers sum

def add(x:int, y:int,)->int:
    c=x+y
    return c
#function calling block

a , b = 10,  20

res= add(a,b)
print(res)

```

```python
# default Arguments:
def add(a,b,c=0, d=0):
    res= a+b+c+d
    return res
print(add (10,20,30,40))
```

```python
#multipliy the  2 or there numbers with default or some other parameters values
def multi(a,b,c=1, d=1,e=1,f=1):
    res= a*b*c*d*e*f
    return res
print(multi(10,20))
```

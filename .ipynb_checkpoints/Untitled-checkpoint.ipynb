```python

#check wether the given numbers are even or odd
num= int(input())
if num>=0 and num % 2 == 0:
    print(f"{num} is +ve Even")
elif num >=0 :
    print(f"{num} is +ve odd")
elif num <=0 and num %2 == 0:
    print(f"{num} is -ve Even")
elif num <0:
    print(f"{num} is -ve odd")
    
else:
    print("zero")
```

```python
#find the smallest number among the three numbers
n1 = int(input())
n2 = int(input())
n3 = int(input())

if n1 < n2 and n1 < n3:
    print("n1 is smaller")
elif n2 < n1 and n2 < n3:
    print("n2 is smaller")
else:
    print("n3 is smaller")
```

```python
year = int(input("enter the year:"))
if year % 400 == 0:
    print("Leap year")
elif year % 100==0:
    print("not a leap year")
elif year % 4==0:
    print("its a leap year")
else:
    print("not a leap year")
    
```

```python
#Largest with equal numbers
a = int(input())
b = int(input())
c = int(input())

if a == b == c:
    print("All numbers are equal")
elif a >= b and a >= c:
    print("Largest =", a)
elif b >= a and b >= c:
    print("Largest =", b)
else:
    print("Largest =", c)
```

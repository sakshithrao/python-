```python
# even , odd or zero

n=int(input())
if n==0:
    print("zero")
elif n% 2 == 0:
    print("even")
else:
    print("odd")
    
```

```python
# positive, nagative or zeron=int(input())
if n>0:
    print("+ve")
elif n<0:
    print("-ve")
else:
    print("zero")
```

```python
# greatest of three numbers
a= int(input())
b= int(input())
c=int(input())

if a>=b and a>=c:
    print("a is greater")
elif b>=a and b>=c:
    print("b is greater")
else:
    print("c is greater")
```

```python
# smallest of 3 numbers 
a= int(input())
b= int(input())
c= int(input())

if a<=b and a<=c:
    print("a  is lower")
elif b<=a and b>c:
    print("b is lower")
else:
    print("c is lower")


```

```python
#Grade Calculator 

marks = int(input())

if marks >= 90:
    print("A")
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
elif marks >= 60:
    print("D")
else:
    print("Fail")
```

```python
#Leap Year (Century Rule)

year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 100 == 0:
    print("Not Leap Year")
elif year % 4 == 0:
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
#access specififiers
# integer -%d
#float - %f
# string - %s
name="sakshith"
age=20
marks=[85,98,65,85,78,98]
#my name is sakshith and i am a 20 years old
print("my name  %s and i am a %d years old"%(name, age))
print("my name  %s and i am a %d years old"%(age,name))

```

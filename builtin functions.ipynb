```python
# sorted(iterable, reverse=false, key=none):
# sort the fruits based on the length of fruit name

fruits=['mango','banana','kivi','apple','water melon']

res =sorted(fruits)
print(res)
print(sorted(fruits, reverse= True))
print(sorted (fruits,reverse=True,key=len))#it sorts the list based on each item length  of each 
```

```python
# map(function, iterable)
#find the square of all list of integers using map function

def square(x):
    return x**2
arr = [1,2,3,4,5]
res=list(map(square,arr))
print(res)
#filter(function, iterable)
```

```python
#filter(function, iterable)
# find even numbers from list of numbers
3 
def check_even(x):
    if x % 2==0:
        return x
lst=[1,2,3,4,5,6]
res=list(filter(check_even,lst))
print(res)
```

```python
# 3 find the 3's  multipliers from list of integers

arr=[1,24,54,22,44,51,33,15]

def  Threes_multipliers(x):
    if x % 3 ==0:
        return 1
res= list(filter(Threes_multipliers,arr))
print(res)
```

```python
#  enumerate(iterable)
# this function returns the list of tuple of index and item for iterable items
fruits = ['Mango','Apple','banana','kivi','water melon']
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f"{ind}-{val} and length is {len(val)}")
```

```python
# zip(iterable 1, iterable2, ...):It returns all itarables as single iterable of tuples

rool_no=[101,102,103,104,105]
names=['ravi','raju','shiva','rani','gopi']
percentage=[90,67,96.8,67.65,99.99999]
print(list(zip(rool_no,names,percentage)))
```

```python
#without all function
percentage=[56,60,70,60,90]
for  num in percentage:
    if num<35:
        print("all students not passed")
        break
else:
    print("all students are pass")
```

```python
# using all function 
percentage=[56,60,70,60,90]
lst=[]
for  num in percentage:
    if num <35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("All students pass")
else:
    print("all students fail")
```

```python
#Squere shaped '*' pattern

n=5
for i in range(n):
    for j in range(n):
        print("*",end="  ")
    print()
```

```python
n=5

for i in range(n):
    for j in range(n):
        print(n* i + j + 1,end=" \t")
    print()
```

```python
n=10
for i in range(n):
    for j in range(n):
        print(j+1,end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(i+j%2,end="\t ")
    print()
```

```python
n=int(input())
k=1
for i in range(n):
    for j in range(n):
        if i >=j:
            print(k, end =" ")
            k-=1
        else:
            print(" " , end ="\t")
    print()
```

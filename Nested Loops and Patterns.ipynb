```python
#Squere shaped '*' pattern

n=5
for i in range(n):
    for j in range(n):
        print("*",end="  ")
    print()
```

```python
n=5

for i in range(n):
    for j in range(n):
        print(n* i + j + 1,end=" \t")
    print()
```

```python
n=10
for i in range(n):
    for j in range(n):
        print(j+1,end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(i+j%2,end="\t ")
    print()
```

```python
n=int(input())
k=1
for i in range(n):
    for j in range(n):
        if i >=j:
            print(k, end =" ")
            k-=1
        else:
            print(" " , end ="\t")
    print()
```

```python
#square patterns

n=6


for row in range(n):
    for col in range(n):
        print("*",end=" ")
    print()
    
```

```python
# rectangle 

rows=4
cols=6

for r in range(rows):
    for c in range(cols):
        print("*",end=" ")
    print()
```

```python
n=5

for r in range(1,n+1):
    for c in range(r):
        print("*",end=" ")
    print()

```

```python
n=5

for r in range(1,n+1):
    for c in range(n-r):
        print(n,end=" ")
    print()
```

```python
n=5

for i in range(n):
    for j in range(n):
        if i==j:
            print("X",end=" ")
        else:
            print("*",end=" ")
    print()
```

```python
n=5

for i in range(n):
    for j in range(n):
        if i+j==n-1:
            print("X",end=" ")
        else:
            print("*",end=" ")
    print()
```

```python
#functions basic idea problems


# with  parameters & with return
# sum of 2 numbers

#function defination block

def add(x:int,y:int)->int:
    c= x+y
    return c
# calling function 

a,b=5,6

res = add(a,b)
print(res)
print(add(1,2))# wec can change variables to get different values
print(add(2,2))

```

```python
#with out parameters and with return 

def add()->int:
    x,y=10,20
    c= x+y
    return c
# calling function 

res = add()
print(res)
print(add())
```

```python
# with parameters without  return 
#when function not returns any value or function doesn't contains return statement
#in this case, function by defaultly it returns 'none' as a result to function call back

def add(x:int,y:int)->int:
    c= x+y
    print("this is from function defination block:",c)
# calling function 


res = add(4,5)
print("this is from function callinng block:",res)
```
